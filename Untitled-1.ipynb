```python
import numpy as np
import matplotlib.pyplot as plt
from decorations import*
load_plot_settings()
```

```python
import astropy.constants as const
import astropy.units as u
```

```python
# constants
k_B= (const.k_B).to(u.erg/u.K).value
m_p = (const.m_p).to(u.g).value
c = (const.c).to(u.cm/u.s).value

print("Constants used")
print('\tk_B =', k_B )
print('\tm_p =', m_p )
print('\tc =', c )

# initial conditions
rho = 2e5 #* u.g /u.cm**3
temp_0 = 1.5e8 #* u.K
Y = 1 

# Calculate non-degenerate temperature threshold
# rho = 10^-8 * T^(3/2)
T_nondegen = (rho * 1e8)**(2/3) # Temperature at which material becomes non-degenerate


print(f"Initial conditions:")
print(f"\tDensity: {rho:.2e} g/cm^3")
print(f"\tTemperature: {temp_0:.2e} K")
print(f"\tNon-degenerate threshold: {T_nondegen:.2e} K")
```

```python
def compute_epsilon3alpha(rho, temp, Y):
    '''energy generation rate per unit mass for heleium flash reaction
    rho : density g/cm^3
    temp : temperature kelvin
    Y : Helium Mass Fraction (1 in Helium core)
    '''
    T9 = temp/(1e9)
    epsilon_3alpha = 5.1e8 * rho **2 * Y**3 / T9**3 * np.exp(-4.4027/T9)
    return epsilon_3alpha  # erg/s/g
    

def compute_cV_e(rho, temp):
    """Calculate electron specific heat (degenerate)"""
    x = compute_x(rho)
    cV_e = (1.35e5 / rho) * temp * x * (1 + x**2)**0.5
    return cV_e # erg/g/K

def compute_x(rho):
    '''normalized fermi momentum
    Computted with rho ~ 2e6 x^3 g/cm^3
    '''
    x = (rho/2e6)**(1/3)
    return x

def compute_cV_He():
    """Calculate Helium ion specific heat for an ideal gas"""
    cV_He = 3 * k_B / (4 * m_p)
    return cV_He

def dT_dt(T, rho, Y):
    """Calculate dT/dt in K/s = epsilon / cV"""
    epsilon = compute_epsilon3alpha(rho, T, Y)
    cV_e = compute_cV_e(rho, T)
    cV_He = compute_cV_He()
    cV_total = cV_e + cV_He
    return epsilon / cV_total
# Runge-Kutta solver
def rk_step(T, rho, Y, dt):
    """Single RK step for new Temperature"""
    k1 = dT_dt(T, rho, Y) * dt
    k2 = dT_dt(T + 0.5*k1, rho, Y) * dt
    k3 = dT_dt(T + 0.5*k2, rho, Y) * dt
    k4 = dT_dt(T + k3, rho, Y) * dt
    T_new = T + (k1/6 + k2/3 + k3/3 + k4/6)
    return T_new
```

```python
# Time integration
T = temp_0
t = 0.0
dt = 1e-3 # seconds
# Storage arrays
times = [t]
temps = [T]

print(f"\nStarting integration...")
step_count = 0

while T <= T_nondegen:
    # Take RK4 step
    T_new = rk_step(T, rho, Y, dt)    
    t += dt
    T = T_new
    # Store data 
    times.append(t)
    temps.append(T)
    step_count += 1

times = np.array(times)
temps = np.array(temps)
times_days = times / 86400  # Convert seconds to days

print(f"\nIntegration complete!")
print(f"  Final time: {times_days[-1]:.6f} days")
print(f"  Final temperature: {temps[-1]:.3e} K")
print(f"  Total steps: {step_count}")
```

```python
fig, ax = plt.subplots(figsize = (7,3.5))
ax.plot(times_days, temps, lw = 1.5, color = 'k')
ax.set_xlabel('Time [days]')
ax.set_ylabel('T [K]')
ax.set_ylim(None,T_nondegen)
plt.show()
fig, ax = plt.subplots(figsize = (7,3.5))
ax.plot(times_days, np.log10(temps), lw = 1.5, color = 'k')
ax.set_xlabel('Time [days]')
ax.set_ylabel('log T /[K]')
ax.set_ylim(None, np.log10(T_nondegen))
```
